````python
A = 0.8
f0 = 120
fs = 44100
phi = np.pi/2
t = np.arange(0, .1, 1.0/fs)
x = A * np.exp(1j*(2*f0*np.pi*t + phi))
````

This means that it is formed by a sinusoid of the fundamental frequency and multiples of it. For instance, we can define the following signal formed by the fundamental frequency plus the second and the third harmonic (note that each wave has a different phase).



````python
f0 = 120
fs = 44100
phi = np.pi/2
t = np.arange(0, .1, 1.0/fs)
x = 1.1*np.cos(2*np.pi*f0*t + np.pi/2) + 0.8*np.cos(2*np.pi*2*f0*t) + 0.2*np.cos(2*np.pi*3*f0*t)
plot_signals(x, fs)
````

3.2. Now, define a harmonic signal, `y` whose fundamental frequency is the defined in Lab 1. Go step by step adding a new harmonic in each step. Plot both signals (the reference and the synthesized) and try to reproduce the shape of the reference signal. 

**Note 1**: in order to have a similar shape, we need to select the phases carefully. One way to do this is to define the harmonic signal as follows:

$$y(t) = \sum_{k=1}^K A_k\cos\left(2\pi kf_0 t + k \phi - (k-1)\pi/2 \right), $$

where $K$ is the number of harmonics, $f_0$ is the fundamental frequency, $A_k$ is the amplitude (weight) of each harmonic and $\phi$ is the phase of the signal (defined in Lab1 Ex 2.4).

**Note 2**:
Define the $A_k$ values relative to the fundamental frequency. This means to define $A_1=1$ and the others less than 1.

**Note 3**: Normalize the amplitude of the signal by the same amplitude of the reference. For instance, if the amplitude of the reference signal is 0.33, you can normalize the syntesized signal by first dividing by its maximum and then multiplyng by 0.33:

```
y = 0.33 * y / np.amax(y)
```

````python
#Parameters
Amax = 0.1117 #Data extracted from the graph
Amin = -0.08465 #Data extracted from the graph
dt = 1.004338 - 1.003 #Data extracted from the graph
T0 = 0.0034  #Data extracted from the graph

A_dc = (Amax-Amin)/2+ (Amin) #Here we are "manually" computing the amplitude of the DC component. 
Ak = [A_dc, 1, 0.95, 0.9, 0.85, 0.8, 0.75]
f0 = 1/T0

fs = 44100
phi = -np.pi*f0*2*dt
t = np.arange(0, 3, 1.0/fs)
x = 0


for k in range(1,7):
  x = x + Ak[k] * np.exp(1j*(2*f0*k*np.pi*t + k*phi - (k-1)*np.pi/2))
  x_plot = A_dc + x
  x_plot = Amax * x_plot / np.amax(x_plot)

  signal_name = 'Synthesized Signal with ' + str(k) + ' harmonics'
  plot_signals([ref, np.real(x_plot)], fs, name=[ 'Reference Signal',signal_name], t_start=1.0, t_end = 1+9*T0)
````
